```python
import numpy as np  # for mathematical and arrays works
import pandas as pd # for using any formats files to work in data frame
import matplotlib.pyplot as mat  # for charts and visualizing data
%matplotlib inline
import seaborn as sb # for charts and visualization
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
# to avoid encoding error, use 'unicode_escape'
```

```python
df.shape #rows and column
```

```python
df.head(10)
```

```python
#Start with cleaning process:
df.info()
```

```python
# if coulumn is no need then kindly drop unrelated/blank columns, axis means full rows, inplace means save and update 
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.info()
```

```python
#null checking process:
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
#drop null values(na)
df.dropna(inplace=True)
```

```python
df.shape
```

```python
pd.isnull(df).sum()
```

```python
#change the data type of amount from float to int
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
#check particular column data type:
df['Amount'].dtypes
```

```python
# check all columns names:
df.columns
```

```python
#if we want to rename column name:
df.rename(columns={'Marital_Status':'Marraige_Status'})
```

```python
#describe() method return description of the data in Mathemattically in the DataFrame(i.e. count, mean, std, etc )
df.describe()
```

```python
#use describe() for specific columns
df[['Age', 'Orders', 'Amount']].describe()
```

```python
#after completion of Data cleaning we will start with EDA(Exploratory Data Analysis)
#Gender
df.columns
```

```python
ax = sb.countplot(x='Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.barplot(x='Gender', y='Amount', data=sales_gen)
```

```python
#From the above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
#AGE:
df.columns
```

```python
ax = sb.countplot(data=df, x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs total age
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.barplot(x='Age Group', y='Amount', data=sales_age)
```

```python
#From the above graph we can notice that most of the buyers are of age group between 26-35 females
```

```python
#State
df.columns
```

```python
#total number of orders state wise:
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

sb.set(rc={'figure.figsize':(25,5)})

sb.barplot(x='State', y='Orders', data=sales_state)
```

```python
#total amount spended state wise:
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(25,5)})

sb.barplot(x='State', y='Amount', data=sales_state)
```

```python
# from the above graphs we can observed total number of sales is equal total number of amount spended (purchasing power) both are equal to each other first position state wise as UP and second as Maharashtra and third as karnataka
```

```python
#Marital Status:
df.columns
```

```python
ax = sb.countplot(data = df, x='Marital_Status')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total number of amounts Marital wise:
sales_marital = df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(6,5)})

sb.barplot(x='Marital_Status', y='Amount',hue='Gender', data=sales_marital)
```

```python
#0=married, 1=singles, from the abpve graph we can see that most of the buyers are married women and they have high purchasing power.
```

```python
#Occupation wise:
df.columns
```

```python
sb.set(rc={'figure.figsize':(20,5)})
ax=sb.countplot(data=df,x='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total number of amount occupation wise:
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(25,5)})

sb.barplot(x='Occupation', y='Amount',data=sales_occupation)
```

```python
#total number of Orders as per Occupation wise and gender:
sales_occupation = df.groupby(['Occupation','Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Occupation', y='Orders',hue='Gender', data=sales_occupation)
```

```python
#total number of Amount spents Occupation wise and gender:
sales_occupation = df.groupby(['Occupation','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Occupation', y='Amount',hue='Gender', data=sales_occupation)
```

```python
#total number of Amount spents Occupation wise :
sales_occupation = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Occupation', y='Amount', data=sales_occupation)
```

```python
#From the above the graph we can notice that the most purachasing power and most order are done by IT Sectors, Healthcare and Aviation of Women only.
```

```python
#Analysis on Product Category:
df.columns
```

```python
sb.set(rc={'figure.figsize':(26,5)})
ax=sb.countplot(data=df,x='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total number of Orders Product wise :
sales_products = df.groupby(['Product_Category'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Product_Category', y='Orders', data=sales_products)
```

```python
#total number of Orders Product wise and Gender :
sales_products = df.groupby(['Product_Category', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Product_Category', y='Orders',hue='Gender', data=sales_products)
```

```python
#total number of amounts spent in Products wise :
sales_products = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Product_Category', y='Amount', data=sales_products)
```

```python
#total number of Orders Product wise and Gender :
sales_products = df.groupby(['Product_Category', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Product_Category', y='Amount',hue='Gender', data=sales_products)
```

```python
#From the above graph we can see most of the sold products are from food, cloths and electronics
```

```python
#checking top selling products using product id:
sales_products_id = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)

sb.set(rc={'figure.figsize':(20,5)})

sb.barplot(x='Product_ID', y='Orders', data=sales_products_id)
```
